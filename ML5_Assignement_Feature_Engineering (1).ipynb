```python
# Q1. What is data encoding? How is it useful in data science?
Ans.
Data encoding is the process of converting categorical data into numerical values so that it can be used by machine
learning algorithms. Machine learning algorithms can only understand numerical data, so categorical data must be 
encoded before it can be used.

Here are some of the benefits of data encoding in data science:
a. It allows machine learning models to understand categorical data.
b. It can improve the accuracy of machine learning models.
c. It can make data analysis easier.
d. It can help to reduce the size of the data set.
```

```python
# Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.
Ans.
Nominal encoding is a type of data encoding that is used to represent categorical data. Categorical data is data that
can be categorized into different groups, such as gender, country, or product type. Nominal encoding assigns a unique
integer value to each category, but the values do not have any order or ranking.
```

```python
# Example of nomial encoding
import seaborn as sns
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoder_sparse = encoder.fit_transform(df[['day','time']])
```

```python
import pandas as pd
encoder_binary = pd.DataFrame(encoder_sparse.toarray(),columns= encoder.get_feature_names_out())
```

```python
encoder_binary 
```

```python
pd.concat([df,encoder_binary],axis=1)
```

```python
# Q3. In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.
Ans.
Nominal encoding and one-hot encoding are two popular techniques for encoding categorical data. Nominal encoding assigns a
unique integer value to each category, while one-hot encoding creates a new binary variable for each category.
```

```python
# Example 
import pandas as pd

df = pd.DataFrame({'Product' : ['A','B','C','C','A','B'],
                   'Rating' : ['High','Medium','Low','Low','High','Medium']})
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encode = OneHotEncoder()
```

```python
encode_sparse = encode.fit_transform(df[['Product']])
```

```python
import pandas as pd
pd.DataFrame(encode_sparse.toarray(),columns= encode.get_feature_names_out())
```

```python
# Q4. Suppose you have a dataset containing categorical data with 5 unique values. Which encoding
# technique would you use to transform this data into a format suitable for machine learning algorithms?
# Explain why you made this choice.
Ans.
If I have a dataset containing categorical data with 5 unique values, I would use one-hot encoding to transform this data into
a format suitable for machine learning algorithms.
One-hot encoding is a technique that creates a new binary feature for each unique value in the categorical variable. For 
example, if the categorical variable has 5 unique values, then one-hot encoding will create 5 new binary features. Each feature
will be 0 or 1, depending on whether the categorical variable has that value or not.
```

```python
# Q6. You are working with a dataset containing information about different types of animals, including their
# species, habitat, and diet. Which encoding technique would you use to transform the categorical data into
# a format suitable for machine learning algorithms? Justify your answer.
Ans.
The encoding technique I would use to transform the categorical data into a format suitable for machine learning algorithms
is one-hot encoding. One-hot encoding is a technique for encoding categorical data into numerical data by creating a binary
vector for each category. Each vector has a 1 in the position corresponding to the category and 0s in all other positions. 
This allows machine learning algorithms to understand the categorical data as a set of independent features.
```
